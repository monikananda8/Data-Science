## Removing Stopwords

```python
from nltk.corpus import stopwords
```

```python
stop= stopwords.words('english')
```

```python
print(stop)
```

```python
Roosevelt_token = regexp_tokenize(Roosvelt.lower(), "[\w']+")
```

```python
filtered_Roosevelt = [word for word in Roosevelt_token if not word in stop]
```

```python
Kennedy_token = regexp_tokenize(Kennedy.lower(), "[\w']+")
filtered_Kennedy = [word for word in Kennedy_token if not word in stop]
```

```python
Nixon_token = regexp_tokenize(Nixon.lower(), "[\w']+")
filtered_Nixon = [word for word in Nixon_token if not word in stop]
```

## Identify Most Common Words

```python
Roosevelt_freq = nltk.FreqDist(filtered_Roosevelt)
#for key,val in Roosevelt_freq.items():
#    print(str(key) + ':' + str(val))
Roosevelt_freq.plot(20, cumulative=False)
```

```python
Roosevelt_freq.most_common(5)
```

```python
Kennedy_freq = nltk.FreqDist(filtered_Kennedy)
Kennedy_freq.plot(20, cumulative=False)
```

```python
Kennedy_freq.most_common(5)
```

```python
Nixon_freq = nltk.FreqDist(filtered_Nixon)
Nixon_freq.plot(20, cumulative=False)
```

```python
Nixon_freq.most_common(5)
```

## Word Cloud Generation

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
wc = WordCloud(background_color='white', relative_scaling=1)
img = wc.generate_from_text(' '.join(filtered_Roosevelt))

# Display the generated image:
plt.imshow(img, interpolation='bilinear')
plt.axis("off")
plt.show()
```
