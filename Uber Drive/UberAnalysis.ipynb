## Q19. Print the miles travelled by the Uber drivers on each day in the month of January. (2 points)

## Q25. Make the following ranges of miles ["<5","5-10","10-20","20-35",">35"]. Keep these in the horizontal axis and count of trips in the vertical axis. Plot a bar plot. (2 points)

```python
df['MILES*'].max()
```

```python
df['MILESRANGE']= pd.cut(df["MILES*"],
       bins=[0,5, 10, 20, 35, 311], 
       labels=["<5", "5-10", "10-20", "20-35", ">35"])
```

```python
sns.countplot(x="MILESRANGE", data=df)
plt.xlabel("Range of Miles")
plt.ylabel("Number of Trips")
```
